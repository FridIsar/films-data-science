```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from random import uniform

df = pd.read_csv('IMDb_movies.csv', low_memory=False)
df
```

```python
indexNames = df[(df['country'] != 'France')].index
df.drop(indexNames , inplace=True)
indexNames2 = df[(df['avg_vote'] != 'NaN')].index
df.drop(indexNames2)
df
```

```python
del df["imdb_title_id"]
# del df["title"]
del df["original_title"]
del df["date_published"]
del df["country"]
del df["language"]
#del df["year"]
del df["writer"]
del df["director"]
del df["actors"]
del df["description"]
#del df["avg_vote"]
del df["votes"]
del df["budget"]
del df["usa_gross_income"]
del df["worlwide_gross_income"]
del df["reviews_from_users"]
del df["reviews_from_critics"]
del df["production_company"]
#del df["duration"]
del df["metascore"]
del df["genre"]
df.dropna()    
df
```

```python
df2 = df

X = df2.drop(columns=['title'])
y = df2['title']
df2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)

```

```python

model.score(X_train, y_train)
```
